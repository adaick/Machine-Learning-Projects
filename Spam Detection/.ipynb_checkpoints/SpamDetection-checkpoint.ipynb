# Spam Detection

### This notebook is used to build a prediction model that will accurately classify which texts are spam or not.

## Data Set

### This data set is taken from Kaggle :- https://www.kaggle.com/uciml/sms-spam-collection-dataset

The SMS Spam Collection is a set of SMS tagged messages that have been collected for SMS Spam research. It contains one set of SMS messages in English of 5,574 messages, tagged acording being ham (legitimate) or spam.

The files contain one message per line. Each line is composed by two columns: v1 contains the label (ham or spam) and v2 contains the raw text.

#### Tokenization
Converts entire message into individual words.

```python
df['token_text'] = df['Text'].apply(lambda text : text.split())
df['token_text']
```

#### Removal of Stopwords

```python
nltk.download('stopwords')
from nltk.corpus import stopwords


stop_words = stopwords.words('english')

# def remove_stopwords(token_text):
#     words=[]
#     for word in token_text:
#         if word not in stop_words:
#             words.append(word)
            
#     return words
```

```python
df['token_text'] = df['token_text'].apply(lambda token_text:[word for word in token_text if word not in stop_words])
df['token_text']
```

### Text Normalization

#### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

wl=WordNetLemmatizer()
```

```python
df['lemma_text'] = df['token_text'].apply(lambda token_text : [wl.lemmatize(word,pos='v') for word in token_text])
df['lemma_text']
```

#### Rejoining individual word back into sentences

```python
df['cleaned_text'] = df['lemma_text'].apply(lambda text: ' '.join(text))
df
```

### Text Vector Generation

#### TF-IDF

```python

from sklearn.feature_extraction.text import TfidfVectorizer

review_vectorizer = TfidfVectorizer()
review_features = review_vectorizer.fit_transform(df['cleaned_text'])

review_features.shape
```

### Model Building

#### Creating Train and Test datsets

```python
X = review_features    #### Features
y = df['num_label']  #### Target

#Using the train_test_split to create train and test sets.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 47, test_size = 0.25)

print('Training set :', X_train.shape)
print('Testing set :', X_test.shape)
```

#### Naive Bayes Model

```python
# import and instantiate a Multinomial Naive Bayes model
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()

#Training the naive bayes classifier.
nb.fit(X_train, y_train)

# make class predictions for X_test_dtm
y_pred_nb = nb.predict(X_test)

# calculate accuracy of class predictions
from sklearn import metrics
print('Accuracy Score on test data: ',metrics.accuracy_score(y_test, y_pred_nb))
```

#### Decision Tree

```python
#Importing the Decision tree classifier from the sklearn library.
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion = 'entropy')

#Training the decision tree classifier. 
clf.fit(X_train, y_train)

#Predicting labels on the test set.
y_pred =  clf.predict(X_test)

#Importing the accuracy metric from sklearn.metrics library

from sklearn.metrics import accuracy_score
print('Accuracy Score on train data: ', accuracy_score(y_true=y_train, y_pred=clf.predict(X_train)))
print('Accuracy Score on test data: ', accuracy_score(y_true=y_test, y_pred=y_pred))
```

#### Random Forests

```python
#Importing the random forest classifier from the sklearn library.
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier()

#Training the random forest classifier.
rf_classifier.fit(X_train,y_train)

#Predicting labels on the test set
y_pred_random =  rf_classifier.predict(X_test)

#Importing the accuracy metric from sklearn.metrics library
from sklearn.metrics import accuracy_score
print('Accuracy Score on train data: ', accuracy_score(y_true=y_train, y_pred=rf_classifier.predict(X_train)))
print('Accuracy Score on test data: ', accuracy_score(y_true=y_test, y_pred=y_pred_random))
```

### Hyperparameter Tuning

#### Random Search
Technique where random ombinations of the hyperparameter are used to find best solution.

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
    'n_estimators' : randint(low=1, high = 200),
    'max_features' : randint(low=1, high = 8)
}

rf_tune = RandomForestClassifier()

#Training the random search
rnd_search = RandomizedSearchCV(rf_tune,param_distributions = param_distribs)
rnd_search.fit(X_train,y_train)

# The best hypermeter combinations
# print("The best hypermeter combinations : ")
rnd_search.best_params_

#Predicting labels on the test set
rnd_pred = rnd_search.predict(X_test)

#Importing the accuracy metric from sklearn.metrics library
from sklearn.metrics import accuracy_score
acc =accuracy_score(y_test,rnd_pred)
acc
```
